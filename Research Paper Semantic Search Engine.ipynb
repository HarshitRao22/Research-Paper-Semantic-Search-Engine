### Explanation of FAISS Usage

FAISS (Facebook AI Similarity Search) is a library that allows you to quickly search for similarity between vectors. In this notebook, it's used to find research papers whose embeddings are most similar to a given query embedding.

```python
faiss.normalize_L2(embedding.reshape(1,-1))
```

#### `faiss.normalize_L2(embedding.reshape(1,-1))`

This line normalizes a vector (in this case, a single `embedding`) to have a Euclidean norm (L2 norm) of 1. Normalization is crucial for cosine similarity (which is effectively what inner product search becomes after L2 normalization) because it ensures that the magnitude of the vectors doesn't influence the similarity score, only their direction. `reshape(1,-1)` is used to make sure the single embedding is treated as a 2D array (a single row of a matrix), which FAISS expects.

```python
index=faiss.IndexFlatIP(384)
```

#### `index=faiss.IndexFlatIP(384)`

This initializes a FAISS index.
*   `IndexFlatIP` creates a simple index that performs an exhaustive search using the inner product (IP) as the similarity metric.
*   `384` is the dimension of the embeddings (the `sample_embedding.shape` and `embedding.shape` showed `(384,)` and `(5, 384)` earlier). The index needs to know the dimension of the vectors it will store and search.

```python
query="deep learning for medical image analysis"
query_embedding=model.encode([query])
query_embedding.shape
```

```python
index.add(embeddings)
print(index.ntotal)
```

#### `index.add(sample_embedding)` and `print(index.ntotal)`

*   `index.add(sample_embedding)` adds your pre-computed embeddings (from `sample_embedding`) to the FAISS index. This is how the index 'learns' about the vectors it needs to search.
*   `index.ntotal` returns the total number of vectors currently stored in the index.

```python
faiss.normalize_L2(query_embedding)
```

#### `faiss.normalize_L2(query_embedding)`

Just like the embeddings stored in the index, the `query_embedding` (the embedding of your search query) also needs to be L2 normalized before performing the similarity search. This ensures consistent comparison with the normalized vectors in the index.

```python
D,I=index.search(query_embedding,5)
print(D)
print(I)
```

#### `D,I=index.search(query_embedding,5)`

This is the core search operation:
*   `index.search()` takes two arguments: the query embedding(s) and `k`, the number of nearest neighbors to retrieve.
*   `query_embedding`: The vector representation of your search query.
*   `5`: Specifies that you want the top 5 most similar items.
*   It returns two arrays:
    *   `D`: An array containing the distances (or similarity scores, as Inner Product is used and normalized vectors make it cosine similarity) of the `k` nearest neighbors to each query.
    *   `I`: An array containing the original indices (from your `df` DataFrame) of the `k` nearest neighbors.

```python
def search_paper(query,k=5):
  query_embedding=model.encode([query])
  faiss.normalize_L2(query_embedding)
  D,I=index.search(query_embedding,k)
  for score,idx in zip(D[0],I[0]):
    print(f"Title: {df.iloc[idx]['title']}")
    print(f"Abstract: {df.iloc[idx]['abstract'][:300]}")
    print(f"Score: {score}")
  return D,I
  print()
```

#### `search_paper(query, k=5)` function

This function encapsulates the entire search process:
1.  Takes a `query` string and an optional `k` (number of results) as input.
2.  Encodes the `query` into an embedding using the `model`.
3.  L2-normalizes the `query_embedding`.
4.  Performs the FAISS `search` to get `D` (scores) and `I` (indices).
5.  Iterates through the top `k` results, using the `idx` to retrieve the `title` and a truncated `abstract` from the original `df` DataFrame.
6.  Prints the title, abstract, and similarity score for each result.
7.  Returns the `D` and `I` arrays.

```python
D,I= search_paper("deep learning for medical image analysis")
print(D)
print(I)
```

```python
def search_and_summarize(query, k=5):

    query_embedding = model.encode([query])
    faiss.normalize_L2(query_embedding)
    D, I = index.search(query_embedding, k)
    all_entities = []
    for score, idx in zip(D[0], I[0]):
        title = df.iloc[idx]["title"]
        abstract = df.iloc[idx]["abstract"]
        print("Title:", title)
        print()
        print("Abstract:")
        print(abstract[:300])
        print()
        print(f"Similarity Score: {score:.3f}")

        if score < 0.20: # Skip if score is below 0.20
            print("Match Level: Too Low. Skipping this paper.\n" + "="*70 + "\n")
            continue

        if score >= 0.70:
            print("Match Level: Highly Similar")
        elif score >= 0.50:
            print("Match Level: Good Match")
        elif score >= 0.30:
            print("Match Level: Moderately Similar")
        else:
            print("Match Level: Weak Match")
        print()
        summary = summarizer(
            abstract,
            max_length=100,
            min_length=30,
            do_sample=False
        )
        print("Summary:")
        print(summary[0]["summary_text"])
        print()
        keywords = kw_model.extract_keywords(
            abstract,
            keyphrase_ngram_range=(1,2),
            stop_words="english",
            top_n=5
        )
        print("Keywords:")
        for keyword, value in keywords:
            print("-", keyword)
        print()
        doc = nlp(abstract)
        print("Named Entities:")
        current_abstract_entities = []
        for entity in doc.ents:
            if entity.text not in current_abstract_entities:
                current_abstract_entities.append(entity.text)
                print(f"- {entity.text} ({entity.label_})")
                if entity.label_ != "CARDINAL" and entity.label_ != "ORDINAL":
                    all_entities.append(entity.text)
        if len(current_abstract_entities) == 0:
            print("No entities found.")
        print("\n" + "=" * 70 + "\n")
    print("Research Insights")
    print("-----------------")
    print("Top 10 Most Common Entities:\n")
    entity_count = {}
    for name in all_entities:
        if name in entity_count:
            entity_count[name] = entity_count[name] + 1
        else:
            entity_count[name] = 1
    sorted_entities = sorted(
        entity_count.items(),
        key=lambda x: x[1],
        reverse=True
    )
    count = 0
    for name, frequency in sorted_entities:
        print(name, ":", frequency)
        count = count + 1
        if count == 10:
            break
    print()
```
